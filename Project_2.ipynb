```python
#remove the invaild values
lux = lux[lux['Unit_Size_Sqft']> 0]
```

```python
lux.info()
```

```python
#fix ticket price cr.
#.loc i use because pandas value dont take value that why i use .loc


lux.loc[:,'Ticket_Price_Cr'] = pd.to_numeric( lux['Ticket_Price_Cr'].astype(str).str.replace("Cr","").str.strip(),errors='coerce')

```

```python
#convert NRI_buyer
lux.loc[:, 'NRI_Buyer'] = lux['NRI_Buyer'].astype(str).str.strip().str.lower()
```

```python
lux.loc  [:, 'NRI_Buyer'] = lux['NRI_Buyer'].map({'yes': 1, 'no' : 0})
```

```python
#convert datatime
lux.loc[:, 'Purchase_Quarter'] = pd.to_datetime(lux['Purchase_Quarter'])
```

```python
lux.info()
```

```python
lux.isnull().sum()
```

```python
#handle missing values

lux_unit = lux['Unit_Size_Sqft'].mean()

lux_unit


```

```python
lux.loc[:,'Unit_Size_Sqft'] = lux['Unit_Size_Sqft'].fillna(lux_unit)
```

```python
lux.describe()
```

```python
lux['NRI_Buyer'].unique()
```

```python
lux['NRI_Buyer'].fillna(0, inplace=True)
```

```python
lux_ticket = lux['Ticket_Price_Cr'].mean()

lux_ticket


```

```python
lux.loc[:,'Ticket_Price_Cr'] = lux['Ticket_Price_Cr'].fillna(lux_ticket)
```

```python
lux.isnull().sum()
```

```python
lux_amenity = lux['Amenity_Score'].mean()

lux_amenity
```

```python
lux.loc[:,'Amenity_Score'] = lux['Amenity_Score'].fillna(lux_amenity)
```
